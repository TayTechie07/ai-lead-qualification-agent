```python

lead_message = input("Paste the lead message here:\n")

def analyze_lead(message):
    keywords = ["need", "support", "ai", "lead", "management", "solution", "agency"]
    score = sum(word in message.lower() for word in keywords)

    if score >= 3:
        intent = "High"
    elif score == 2:
        intent = "Medium"
    else:
        intent = "Low"

    return intent, score

def explain_intent(intent, score):
    return f"The lead shows {intent.lower()} intent based on keyword relevance (score: {score})."


def suggest_reply(intent):
    if intent == "High":
        return "Thank you for reaching out. We’d be happy to discuss how AI can support your lead management process."
    elif intent == "Medium":
        return "Thanks for your message. Please share more details so we can better understand your needs."
    else:
        return "Thank you for your interest. Feel free to reach out if you’d like to explore this further."

intent, score = analyze_lead(lead_message)
reason = explain_intent(intent, score)
reply = suggest_reply(intent)

print("\n----- AI AGENT RESULT -----")
print("Intent Level:", intent)
print("Explanation:", reason)
print("\nSuggested Reply:")
print(reply)
```
